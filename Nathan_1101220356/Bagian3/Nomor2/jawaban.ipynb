```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

  
# read the image 
image = cv2.imread('ayon.jpeg') 
  
# convert color image into grayscale image 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

  
# cmap stands for colormap 
plt.imshow(image, cmap="gray", vmin=0, vmax=255)

filename = 'ayon_grayscale.jpeg'
# Using cv2.imwrite() method 
# Saving the image 
cv2.imwrite(filename, image) 
  
# display that image 
plt.show() 
```

```python

  
# read the image 
image = cv2.imread('ayon.jpeg') 
  
# convert color image into grayscale image 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 


# cmap stands for colormap 
plt.imshow(image, cmap="gray", vmin=0, vmax=255)

(thresh, im_bw) = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
thresh = 127
im_bw = cv2.threshold(image, thresh, 255, cv2.THRESH_BINARY)[1]


filename = 'ayon_bnw.jpeg'
# Using cv2.imwrite() method 
# Saving the image 
cv2.imwrite(filename, im_bw) 
  
# display that image 
plt.show() 
```

```python
fig = plt.figure(figsize=(10, 7)) 

rows = 3
columns = 1

# reading images 
Image1 = cv2.imread('ayon.jpeg')
Image1 = cv2.cvtColor(Image1, cv2.COLOR_RGB2BGR) 
Image2 = cv2.imread('ayon_grayscale.jpeg') 
Image2 = cv2.cvtColor(Image2, cv2.COLOR_RGB2BGR) 
Image3 = cv2.imread('ayon_bnw.jpeg') 
Image3 = cv2.cvtColor(Image3, cv2.COLOR_RGB2BGR) 

# Adds a subplot at the 1st position 
fig.add_subplot(columns, rows, 1) 

# showing image 
plt.imshow(Image1) 
plt.axis('off') 
plt.title("Normal") 

# Adds a subplot at the 2nd position 
fig.add_subplot(columns, rows, 2) 

# showing image 
plt.imshow(Image2) 
plt.axis('off') 
plt.title("Grayscale") 

# Adds a subplot at the 3rd position 
fig.add_subplot(columns, rows, 3) 

# showing image 
plt.imshow(Image3) 
plt.axis('off') 
plt.title("Black and White") 



```
