```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

  
# read the image 
image = cv2.imread('emma.jpg') 
  
# convert color image into grayscale image 
img1 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) 
  
# cmap stands for colormap 
plt.imshow(img1) 
  
# display that image 
plt.show() 
```

```python
fig = plt.figure(figsize=(10, 7)) 

rows = 4
columns = 1

# reading images 
Image1 = cv2.imread('emma.jpg')
Image1 = cv2.cvtColor(Image1, cv2.COLOR_RGB2BGR) 
Image2 = cv2.imread('emma_flip_horizontal.jpg') 
Image2 = cv2.cvtColor(Image2, cv2.COLOR_RGB2BGR) 
Image3 = cv2.imread('emma_flip_vertikal.jpg') 
Image3 = cv2.cvtColor(Image3, cv2.COLOR_RGB2BGR) 
Image4 = cv2.imread('emma_transpose.jpg') 
Image4 = cv2.cvtColor(Image4, cv2.COLOR_RGB2BGR) 

# Adds a subplot at the 1st position 
fig.add_subplot(columns, rows, 1) 

# showing image 
plt.imshow(Image1) 
plt.axis('off') 
plt.title("Normal") 

# Adds a subplot at the 2nd position 
fig.add_subplot(columns, rows, 2) 

# showing image 
plt.imshow(Image2) 
plt.axis('off') 
plt.title("Flip Horizontal") 

# Adds a subplot at the 3rd position 
fig.add_subplot(columns, rows, 3) 

# showing image 
plt.imshow(Image3) 
plt.axis('off') 
plt.title("Flip Vertikal") 

# Adds a subplot at the 4th position 
fig.add_subplot(columns, rows, 4) 

# showing image 
plt.imshow(Image4) 
plt.axis('off') 
plt.title("Transpose") 



```
